5. **How do cumulative distribution functions (CDF) differ from probability distribution functions (PDF)?**
-  The Probability Density Function (PDF) and Cumulative Distribution Function (CDF) are both functions used to describe the distribution of a random variable, but they represent different aspects. The PDF gives the probability density at a specific point, while the CDF gives the probability that a random variable is less than or equal to a specific value.
- **probability distribution functions (PDF)**
  - Defination: The probability density function or PDF describes a continuous random variable’s probability distribution. It shows the probability that the random variable will have a particular value.
  - A continuous random variable is frequently represented using the expression f(x), where ‘x’ represents the variable’s value.
  - The PDF is used for continuous random variables, where the probability is distributed over an infinite range of values.
  - The PDF provides the probability density at a particular point on the continuous distribution curve, indicating how the probability is spread across different values.
  - The integral of the PDF over a certain range yields the probability of the random variable falling within that range.
  - The PDF can take any non-negative value for any given point on the distribution curve, representing the likelihood of the variable assuming that value.
  - The PDF is commonly used in probability density estimation, statistical modelling, and understanding the shape of continuous distributions.

- **cumulative distribution functions (CDF)**
  - Defination:  In general, the probability that a random variable will have a value less than or equal to a specific value is determined by the cumulative distribution function or CDF.
  - It can be applied to continuous and discrete random variables and is frequently expressed as F(x), where ‘x’ represents the variable’s value.
  - The CDF applies to discrete and continuous random variables, as it accumulates probabilities for all possible values of the random variable.
  - The CDF gives the cumulative probability up to a specific value, offering insights into the probabilities of the random variable being less than or equal to that value.
  - The CDF is obtained by integrating the PDF from a lower bound to a specific value, ‘x’, which accumulates the probabilities up to that point.
  - The CDF always ranges from 0 to 1, as it gives the cumulative probability, and it is non-decreasing, meaning it can only increase or remain constant as ‘x’ increases.
  - The CDF can be used to determine a distribution’s percentiles and quantiles and the likelihood that a random variable will fall within a certain range.

7. **What are the key properties of a Bernoulli distribution?**
-  A Bernoulli Distribution is basically the probability that an experiment produces a specific outcome. It is simply a binomial distribution with a single event (that is n = 1).
-  The key properties of a Bernoulli distribution:
  - Binary Outcomes:  The core characteristic is that it deals with events with only two possible outcomes.  These are often labeled as "success" and "failure" or represented numerically as 1 and 0. This makes it useful for modeling yes/no, true/false, or success/failure scenarios.
  - Single Parameter (p):  The Bernoulli distribution is defined by a single parameter, denoted as \(p\). This parameter represents the probability of success in a single Bernoulli trial.
  - Constant Probability: The probability of success (\(p\)) remains the same for every trial.
  - Independent Trials:  Each Bernoulli trial is independent, meaning the outcome of one trial does not influence any subsequent trials.
  - Complementary Probability: The probability of failure (\(q\)) is the complement of the probability of success and is calculated as \(1-p\).  So, the probabilities of success and failure sum to 1 (\(p+q=1\)).
  - Discrete Probability Distribution:  It is a discrete probability distribution because it involves a finite number of distinct outcomes (only 0 or 1).
  - Mean (Expected Value): The expected value or mean of a Bernoulli distribution is equal to the probability of success (\(p\)).  This represents the average outcome in a series of Bernoulli trials.
  - Variance: The variance of a Bernoulli distribution is given by the formula \(p(1-p)\) or \(pq\). It measures how much the outcomes deviate from the mean. The variance is maximized when \(p=0.5\) and decreases as \(p\) approaches 0 or 1, signifying lower uncertainty when outcomes are more predictable.
  - Symmetry/Asymmetry: The shape of the distribution depends on the value of \(p\).
  - Foundation for Other Distributions: The Bernoulli distribution acts as a building block for more complex distributions like the binomial distribution (multiple independent Bernoulli trials) and the geometric distribution (number of trials until the first success).
  - Memoryless Distribution: Some sources state that the Bernoulli distribution is a memoryless distribution.

14. **How does the Central Limit Theorem relate to the normal distribution?**
-   The Central Limit Theorem (CLT) is a fundamental statistical concept that establishes a powerful relationship with the normal distribution, particularly concerning sample means
-   The core of the Central Limit Theorem lies in its statement that as the sample size (n) increases, the distribution of the sample means will approximate a normal distribution, regardless of the shape of the original population distribution.
-   Key Aspects of the Relationship:
  - Convergence to Normality: The CLT states that even if samples are taken from a population with a skewed or non-normal distribution, the distribution of the means of those samples will start to resemble a bell-shaped normal curve as sample sizes increase.
  - Sample Mean Distribution: The CLT focuses on the sampling distribution of the mean, which is the distribution of the means of many samples taken from the same population. This sampling distribution will become approximately normal with a large enough sample size.
  - Mean and Standard Deviation: The CLT also specifies the parameters of this approximating normal distribution:
  
  > The mean of the sampling distribution of the means will be equal to the mean of the original population (μ).
  
  > The standard deviation of the sampling distribution of the means (also known as the standard error of the mean) will be equal to the population standard deviation (σ) divided by the square root of the sample size (√n). Larger sample sizes lead to smaller standard errors and more precise estimations of the population mean.

15. **What is the application of Z statistics in hypothesis testing?**
-   Z-statistics, calculated using a Z-test, are applied in hypothesis testing to determine if there's a significant difference between a sample's mean and a population's mean, or between the means of two independent samples.
-   The z-statistic plays a crucial role in hypothesis testing, particularly when dealing with large sample sizes (typically greater than 30) and known population variance. It helps determine if there is a significant difference between:
-    A sample mean and a known population mean (One-sample Z-test):
  - Application: Determining if a sample, like a group of students' test scores, is significantly different from the national average, given that the national average and its standard deviation are known.
  - Formula: z = (x̄ - μ) / (σ/√n)

   >x̄: sample mean

   >μ: population mean

   >σ: population standard deviation

   >n: sample size
-   The means of two independent samples (Two-sample Z-test):
  - Application: Comparing the average performance of two groups, such as students who took different learning methods, to see if there's a statistically significant difference in their scores.
  - Formula: z = (x̄₁ - x̄₂) - (μ₁ - μ₂) / √((σ₁²/n₁) + (σ₂²/n₂))
   
   >x̄₁, x̄₂: sample means of the two groups
   
   >μ₁, μ₂: population means of the two groups
   
   >σ₁², σ₂²: population variances of the two groups
   
   >n₁, n₂: sample sizes of the two groups
-   The proportion of a characteristic in a sample to a known population proportion (One-proportion Z-test):
  - Application: Determining if the percentage of people who favor a certain product in a survey differs significantly from a known benchmark.
-   The proportions of a characteristic in two independent samples (Two-proportion Z-test):
  - Application: Comparing the percentage of defects in two different production lines to see if there is a significant difference in their quality control procedures.

16. **How do you calculate a Z-score, and what does it represent?**
-   The Z-score, also known as the standard score, is a statistical measurement describing a data point's relationship to the mean of a dataset. It indicates how many standard deviations a raw score is from the mean. The formula for calculating a Z-score is :

 `\(z=\frac{x-\mu }{\sigma }\) `
-   Where:
  - z: The Z-score
  - x: The raw score (the value being evaluated)
  - μ (mu): The mean of the dataset
  - σ (sigma): The standard deviation of the dataset  
-   To calculate it, subtract the mean from the raw score and divide by the standard deviation. For example, a score of 190 on a test with a mean of 150 and a standard deviation of 25 would have a Z-score of 1.6.
-   A positive Z-score means the data point is above the mean, while a negative Z-score means it is below. A Z-score of 0 means the data point equals the mean. The absolute value of the Z-score indicates how far the data point is from the mean.  

17. **What are point estimates and interval estimates in statistics?**
-   The point estimates and interval estimates are two fundamental ways to estimate unknown population parameters using sample data. Point estimates provide a single value as the best guess, while interval estimates provide a range of values within which the parameter is likely to fall.
-   **Point Estimate**
  - A point estimate is a single value derived from sample data that is used as a best guess for an unknown population parameter, such as the mean or proportion. For example, the sample mean is a point estimate for the population mean.
  - While point estimates are easy to calculate, they do not indicate the variability or uncertainty of the estimate

-   **interval estimates**
  - An interval estimate provides a range of values that likely contains the unknown population parameter. This range is often called a confidence interval and is associated with a specific confidence level (like 95%) that reflects the probability that the interval includes the true parameter.
  - Interval estimates are more informative than point estimates because they show the potential variability of the estimate. A key part of an interval estimate is the margin of error, which indicates the range where the true parameter is expected to be.

18. **What is the significance of confidence intervals in statistical analysis?**
-   Confidence intervals in statistical analysis are crucial because they provide a range of plausible values for an unknown population parameter, offering a more informative measure of uncertainty than a single point estimate. They help researchers understand the reliability and precision of their estimates, guiding them in making data-driven decisions.
-   The significance of confidence intervals includes:
  - Quantifying Uncertainty: Instead of reporting a single point estimate, a confidence interval offers a range. This provides a more realistic picture of the uncertainty surrounding that estimate. This helps measure uncertainty in predictions and estimates.
  - Improving Interpretation: Confidence intervals help avoid misinterpretations by showing the possible range of values for a result, rather than just a single number. For example, a 95% confidence interval for the average height of students between 160 cm and 170 cm means that if similar samples are repeatedly taken, 95% of those intervals would contain the true average height of all students in the population.

If a confidence interval for a difference between groups includes zero, it indicates that the results are not statistically significant at the chosen confidence level. For ratios like odds ratios or relative risks, a confidence interval that includes one suggests no statistically significant effect.
  - Building Credibility: By presenting confidence intervals alongside statistical estimates, researchers demonstrate the precision and potential variation around their findings, which builds credibility and transparency.
  - Facilitating Decision-Making: Confidence intervals are used in various fields, such as biomedical research and A/B testing, to help make informed decisions.   - Communicating Results Effectively: Confidence intervals allow researchers to communicate the reliability of their findings by providing a range rather than a single number, which is easier for a wider audience to understand.

19. **What is the relationship between a Z-score and a confidence interval?**
-   In statistics, the Z-score and confidence interval are related because the Z-score is used in the calculation of the confidence interval, particularly when estimating the population mean from a large sample (typically n ≥ 30) with a known population standard deviation.
-   Relationship between Z-scores and Confidence Intervals:
  - Confidence Interval: The Plausible Range: A confidence interval provides a range of values where the true population parameter is likely to be.
  - Z-score: Defining Confidence Boundaries: The Z-score determines the width of the confidence interval.  
-   How the Z-score is used in calculating the confidence interval:
  - Confidence Level and Significance Level: You first choose a desired confidence level (e.g., 90%, 95%, or 99%), according to GeeksforGeeks. The complement of the confidence level is the significance level (α), where α = 1 - Confidence Level. For example, a 95% confidence level corresponds to a 5% significance level (α = 0.05).
  - Finding the Z-score (Zα/2): The significance level is divided by 2 (α/2) to represent the area in each tail of the standard normal distribution. You then use a Z-table or a statistical calculator to find the Z-score that corresponds to the cumulative probability of (1 - α/2). This critical Z-score represents the boundaries that encompass the desired percentage of the distribution.
  - Calculating the Confidence Interval: The formula for a confidence interval for the mean, when the population standard deviation (σ) is known and the sample size is large, is: `x̄ ± Zα/2 * (σ / √n)`
  
  >x̄ is the sample mean.
  
  >Zα/2 is the critical Z-score determined in the previous step.
  
  >σ is the population standard deviation.
  
  >n is the sample size.
  
  - Margin of Error: The term after the ± symbol (Zα/2 * σ/√n) is called the margin of error. It reflects the uncertainty in the estimate and the width of the confidence interval.

21. **What are the assumptions for applying the Central Limit Theorem?**
-   The Central Limit Theorem (CLT) is a fundamental theorem in statistics that states that, given a sufficiently large sample size, the sampling distribution of the sample mean will approximate a normal distribution, regardless of the distribution of the population from which the samples are drawn.
-   Assumptions for applying the Central Limit Theorem:
  - Random Sampling: The samples must be drawn randomly from the population. This means that each member of the population has an equal chance of being selected for the sample.
  - Independence: The samples drawn should be independent of each other. The selection or results from one sample should not influence other samples or their results.
  - Sample Size: The sample size must be "sufficiently large." A common rule of thumb is that a sample size of 30 or more is generally considered sufficient for the CLT to hold, even if the population distribution is not normal. However, if the population is highly skewed, a larger sample size may be required. If the population distribution is already close to normal, smaller sample sizes may be sufficient.
  - Finite Variance: The population from which the samples are drawn must have a finite variance. Distributions with infinite variance (like the Cauchy distribution) are exceptions to the CLT.
  - 10% Condition (for sampling without replacement): When sampling is done without replacement, the sample size should not exceed 10% of the total population. This helps to ensure that the samples remain largely independent.

23. **How does a probability distribution relate to the expected outcome of a random variable?**
-   A probability distribution is a mathematical function that describes the likelihood of different outcomes or values of a random variable. It essentially maps each possible outcome of a random event to its probability of occurring.
-   The expected value of a random variable is a way to quantify the long-term average outcome of that variable based on its probability distribution. It's a measure of the center or "typical" result of an uncertain situation.
-   Here's how they relate:
  - Foundation for Calculation: The probability distribution provides the probabilities for each possible value of a random variable. This information is crucial for calculating the expected value.
  - Weighted Average: The expected value is calculated as a weighted average of all possible values the random variable can take, with the probabilities from the distribution acting as the weights.
  
  >For a discrete random variable X with possible values x and corresponding probabilities P(X=x), the expected value E(X) is calculated as:
  `E(X) = Σ [x * P(X=x)]`
  
  >For a continuous random variable X with probability density function f(x), the expected value E(X) is calculated using integration:
  `E(X) = ∫ x * f(x) dx`
  
  - Long-Term Average: The expected value represents the theoretical average outcome you would expect to observe if you were to repeat the random experiment many times. The law of large numbers states that as the number of trials increases, the average of the results will converge to the expected value.
  - Center of Distribution: The expected value is a measure of the central tendency of the probability distribution. It can be thought of as the point where the distribution would balance if placed on a scale.

3. **Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution.**

```python
def bernoulli_pmf(x, p):
    """
    Calculate the probability mass function (PMF) of a Bernoulli distribution.

    Parameters:
    x (int): The outcome (0 or 1)
    p (float): Probability of success (P(X=1)), where 0 <= p <= 1

    Returns:
    float: PMF value for given x
    """
    if x not in (0, 1):
        return 0.0
    if not (0 <= p <= 1):
        raise ValueError("Probability p must be between 0 and 1.")
    return p if x == 1 else (1 - p)

# Example usage
print("P(X=0):", bernoulli_pmf(0, 0.3))
print("P(X=1):", bernoulli_pmf(1, 0.3))
```

9. **Write a Python function to calculate Z-scores from a dataset and plot them.**

```python
import numpy as np
import matplotlib.pyplot as plt

def calculate_and_plot_zscores(data):
    """
    Calculate Z-scores for the given dataset and plot them.

    Parameters:
    data (array-like): Numeric dataset (list, numpy array, etc.)
    """
    data = np.array(data)
    mean = np.mean(data)
    std = np.std(data, ddof=0)  # Population standard deviation

    # Calculate Z-scores
    z_scores = (data - mean) / std

    # Plot Z-scores
    plt.figure(figsize=(8, 4))
    plt.scatter(range(len(data)), z_scores, color='blue')
    plt.axhline(0, color='red', linestyle='--', label='Mean (Z=0)')
    plt.axhline(1, color='green', linestyle='--', label='Z=1')
    plt.axhline(-1, color='green', linestyle='--', label='Z=-1')
    plt.axhline(2, color='orange', linestyle='--', label='Z=2')
    plt.axhline(-2, color='orange', linestyle='--', label='Z=-2')

    plt.title('Z-scores of the Dataset')
    plt.xlabel('Index')
    plt.ylabel('Z-score')
    plt.legend()
    plt.grid(True)
    plt.show()

    return z_scores

# Example usage:
data_example = [10, 12, 9, 15, 20, 18, 14, 10, 8, 25]
zs = calculate_and_plot_zscores(data_example)
print("Z-scores:", zs)
```

14. **Write a Python program to calculate the Z-score for a given data point and compare it to a standard normal distribution.**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def calculate_z_score(data, x):
    """
    Calculate the Z-score of a data point x given the dataset.
    """
    mean = np.mean(data)
    std = np.std(data, ddof=0)  # Population std deviation
    z = (x - mean) / std
    return z

def plot_standard_normal_with_z(z):
    """
    Plot the standard normal distribution and mark the Z-score on it.
    """
    x = np.linspace(-4, 4, 1000)
    pdf = norm.pdf(x)

    plt.plot(x, pdf, label='Standard Normal PDF')
    plt.fill_between(x, 0, pdf, color='lightblue', alpha=0.5)

    # Mark the Z-score
    z_pdf = norm.pdf(z)
    plt.plot(z, z_pdf, 'ro', label=f'Z-score = {z:.2f}')
    plt.axvline(z, color='r', linestyle='--')

    plt.title('Standard Normal Distribution with Z-score')
    plt.xlabel('Z')
    plt.ylabel('Probability Density')
    plt.legend()
    plt.grid(True)
    plt.show()

# Example usage
data = [10, 12, 9, 15, 20, 18, 14, 10, 8, 25]
x = 18

z = calculate_z_score(data, x)
print(f"Z-score of data point {x}: {z:.2f}")

plot_standard_normal_with_z(z)
```

15. **Implement hypothesis testing using Z-statistics for a sample dataset.**

```python
import numpy as np
from scipy.stats import norm

def z_test(sample, mu_0, alpha=0.05):
    """
    Perform a two-tailed Z-test for population mean.

    Parameters:
    - sample: array-like sample data
    - mu_0: hypothesized population mean (null hypothesis)
    - alpha: significance level (default 0.05)

    Returns:
    - z_stat: calculated Z statistic
    - p_value: two-tailed p-value
    - conclusion: string to interpret the test result
    """
    n = len(sample)
    sample_mean = np.mean(sample)
    sample_std = np.std(sample, ddof=1)  # sample standard deviation
    se = sample_std / np.sqrt(n)          # standard error

    # Calculate Z statistic
    z_stat = (sample_mean - mu_0) / se

    # Calculate two-tailed p-value
    p_value = 2 * (1 - norm.cdf(abs(z_stat)))

    # Decision
    if p_value < alpha:
        conclusion = f"Reject the null hypothesis (p = {p_value:.4f} < {alpha})"
    else:
        conclusion = f"Fail to reject the null hypothesis (p = {p_value:.4f} ≥ {alpha})"

    return z_stat, p_value, conclusion

# Example sample data
sample_data = [52, 55, 53, 58, 54, 50, 49, 57, 56, 54]

# Hypothesized population mean
mu_0 = 50

# Significance level
alpha = 0.05

z_stat, p_val, result = z_test(sample_data, mu_0, alpha)

print(f"Sample Mean: {np.mean(sample_data):.2f}")
print(f"Z-statistic: {z_stat:.4f}")
print(f"P-value: {p_val:.4f}")
print(result)
```

16. **Create a confidence interval for a dataset using Python and interpret the result.**
-  A confidence interval gives a range of values within which we expect the true population mean to lie, with a specified confidence level (e.g., 95%)

```python
import numpy as np
from scipy.stats import norm

def confidence_interval(data, confidence=0.95):
    """
    Calculate the confidence interval for the mean of the dataset.

    Parameters:
    - data: array-like sample data
    - confidence: confidence level (default 0.95 for 95%)

    Returns:
    - (lower_bound, upper_bound): tuple representing the confidence interval
    """
    n = len(data)
    mean = np.mean(data)
    std = np.std(data, ddof=1)  # sample standard deviation
    se = std / np.sqrt(n)        # standard error

    # Z-score for the confidence level
    z = norm.ppf(1 - (1 - confidence) / 2)

    # Calculate margin of error
    margin_error = z * se

    lower_bound = mean - margin_error
    upper_bound = mean + margin_error

    return lower_bound, upper_bound

# Example dataset
data = [52, 55, 53, 58, 54, 50, 49, 57, 56, 54]

# Calculate 95% confidence interval
conf_level = 0.95
lower, upper = confidence_interval(data, confidence=conf_level)

print(f"Sample Mean: {np.mean(data):.2f}")
print(f"{int(conf_level*100)}% Confidence Interval: ({lower:.2f}, {upper:.2f})")
print(f"Interpretation: We are {int(conf_level*100)}% confident that the true population mean lies between {lower:.2f} and {upper:.2f}.")
```

17. **Generate data from a normal distribution, then calculate and interpret the confidence interval for its mean**.

```python
import numpy as np
from scipy.stats import norm

def confidence_interval(data, confidence=0.95):
    n = len(data)
    mean = np.mean(data)
    std = np.std(data, ddof=1)  # sample std dev
    se = std / np.sqrt(n)
    z = norm.ppf(1 - (1 - confidence) / 2)
    margin_error = z * se
    return mean, mean - margin_error, mean + margin_error

# Generate sample data from a normal distribution
np.random.seed(42)  # for reproducibility
mu = 100
sigma = 15
sample_size = 50

data = np.random.normal(loc=mu, scale=sigma, size=sample_size)

# Calculate 95% confidence interval
mean, lower, upper = confidence_interval(data, confidence=0.95)

print(f"Sample Mean: {mean:.2f}")
print(f"95% Confidence Interval: ({lower:.2f}, {upper:.2f})")
print(f"Interpretation: We are 95% confident that the true population mean lies between {lower:.2f} and {upper:.2f}.")
```

24. **Write a Python function to calculate the skewness and kurtosis of a dataset generated from a normal distribution.**

```python
import numpy as np
from scipy.stats import skew, kurtosis

def calc_skewness_kurtosis(mean=0, std=1, size=1000):
    # Generate normal distribution data
    data = np.random.normal(loc=mean, scale=std, size=size)

    # Calculate skewness
    data_skewness = skew(data)

    # Calculate kurtosis (Fisher’s definition: subtract 3 to make normal dist kurtosis = 0)
    data_kurtosis = kurtosis(data)

    return data_skewness, data_kurtosis

# Example usage
skewness_val, kurtosis_val = calc_skewness_kurtosis()
print(f"Skewness: {skewness_val:.4f}")
print(f"Kurtosis: {kurtosis_val:.4f}")
```
